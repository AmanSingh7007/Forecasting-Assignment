```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df.columns = ['state', 'date', 'sales', 'category']

df['date'] = pd.to_datetime(df['date'])

df['sales'] = (
    df['sales']
    .astype(str)
    .str.replace(',', '')
    .astype(float)
)

df = df.sort_values(['state', 'date'])

all_states = df['state'].unique()

final_df = []

for state in all_states:

    state_df = df[df['state'] == state].copy()

    state_df.dropna(subset=['date'], inplace=True)

    state_df = state_df.groupby('date').agg({'sales': 'sum', 'category': 'first'}).reset_index()

    state_df = state_df.set_index('date')

    full_dates = pd.date_range(
        start=state_df.index.min(),
        end=state_df.index.max(),
        freq='W'
    )

    state_df = state_df.reindex(full_dates)

    state_df['state'] = state

    state_df['category'] = state_df['category'].ffill()

    state_df['sales'] = state_df['sales'].interpolate()

    state_df = state_df.reset_index()

    state_df.rename(columns={'index':'date'}, inplace=True)

    final_df.append(state_df)

df = pd.concat(final_df)

# Lag Features
df['lag_1'] = df.groupby('state')['sales'].shift(1)
df['lag_7'] = df.groupby('state')['sales'].shift(7)
df['lag_30'] = df.groupby('state')['sales'].shift(30)

# Rolling Features
df['rolling_mean_4'] = (
    df.groupby('state')['sales']
      .transform(lambda x: x.rolling(4).mean())
)

df['rolling_std_4'] = (
    df.groupby('state')['sales']
      .transform(lambda x: x.rolling(4).std())
)

# Date Features
df['month'] = df['date'].dt.month
df['quarter'] = df['date'].dt.quarter
df['weekofyear'] = df['date'].dt.isocalendar().week.astype(int)

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['category_encoded'] = le.fit_transform(df['category'])

df = df.dropna()

print(df.head())
print(df.shape)


```

```python
sample_state = df['state'].iloc[4]

temp = df[df['state'] == sample_state]

plt.figure(figsize=(15,5))
plt.plot(temp['date'], temp['sales'])

plt.title(f'Sales Trend - {sample_state}')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.show()
```

```python

split_date = df['date'].max() - pd.Timedelta(weeks=8)

train_df = df[df['date'] <= split_date]
test_df = df[df['date'] > split_date]

print("Train Shape:", train_df.shape)
print("Test Shape:", test_df.shape)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np



features = [
    'lag_1',
    'lag_7',
    'lag_30',
    'rolling_mean_4',
    'rolling_std_4',
    'month',
    'quarter',
    'weekofyear',
    'category_encoded'
]

target = 'sales'

X_train = train_df[features]
y_train = train_df[target]

X_test = test_df[features]
y_test = test_df[target]



xgb_model = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)



xgb_model.fit(X_train, y_train)



xgb_pred = xgb_model.predict(X_test)



rmse = np.sqrt(mean_squared_error(y_test, xgb_pred))
mae = mean_absolute_error(y_test, xgb_pred)

print("XGBoost RMSE:", rmse)
print("XGBoost MAE:", mae)
```

```python
from prophet import Prophet


prophet_train = train_df[['date', 'sales']]

prophet_train.columns = ['ds', 'y']



prophet_model = Prophet()

prophet_model.fit(prophet_train)



future = prophet_model.make_future_dataframe(periods=8, freq='W')

forecast = prophet_model.predict(future)



prophet_pred = forecast.tail(8)['yhat'].values



actual = test_df.groupby('date')['sales'].sum().values[:8]


rmse_prophet = np.sqrt(mean_squared_error(actual, prophet_pred))

print("Prophet RMSE:", rmse_prophet)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX


train_series = train_df.groupby('date')['sales'].sum()

test_series = test_df.groupby('date')['sales'].sum()



sarima_model = SARIMAX(
    train_series,
    order=(1,1,1),
    seasonal_order=(1,1,1,12)
)

sarima_result = sarima_model.fit()



sarima_pred = sarima_result.forecast(steps=8)


rmse_sarima = np.sqrt(
    mean_squared_error(test_series.values[:8], sarima_pred)
)

print("SARIMA RMSE:", rmse_sarima)
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout


series = df.groupby('date')['sales'].sum().values.reshape(-1,1)


scaler = MinMaxScaler()

scaled = scaler.fit_transform(series)


X = []
y = []

window = 8

for i in range(window, len(scaled)):
    X.append(scaled[i-window:i])
    y.append(scaled[i])

X = np.array(X)
y = np.array(y)



split = int(len(X)*0.8)

X_train = X[:split]
X_test = X[split:]

y_train = y[:split]
y_test = y[split:]


lstm_model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_train.shape[1],1)),
    Dropout(0.2),

    LSTM(32),

    Dense(1)
])

lstm_model.compile(
    optimizer='adam',
    loss='mse'
)

history = lstm_model.fit(
    X_train,
    y_train,
    epochs=50,
    batch_size=16,
    validation_data=(X_test, y_test),
    verbose=1
)



lstm_pred = lstm_model.predict(X_test)



lstm_pred = scaler.inverse_transform(lstm_pred)
y_test_actual = scaler.inverse_transform(y_test)



rmse_lstm = np.sqrt(
    mean_squared_error(y_test_actual, lstm_pred)
)

print("LSTM RMSE:", rmse_lstm)
```

```python
results = pd.DataFrame({

    'Model': [
        'XGBoost',
        'Prophet',
        'SARIMA',
        'LSTM'
    ],

    'RMSE': [
        rmse,
        rmse_prophet,
        rmse_sarima,
        rmse_lstm
    ]
})


results = results.sort_values('RMSE')

print(results)
```

```python
best_model_name = results.iloc[0]['Model']

print("Best Model:", best_model_name)
```

```python


future_predictions = []

last_data = df.copy()

future_dates = pd.date_range(
    start=df['date'].max() + pd.Timedelta(weeks=1),
    periods=8,
    freq='W'
)

states = df['state'].unique()

for state in states:

    state_df = last_data[last_data['state'] == state].copy()

    for future_date in future_dates:

        last_row = state_df.iloc[-1]

        new_row = {}

        new_row['date'] = future_date
        new_row['state'] = state
        new_row['category'] = last_row['category']


        new_row['lag_1'] = last_row['sales']

        new_row['lag_7'] = (
            state_df.iloc[-7]['sales']
            if len(state_df) >= 7 else last_row['sales']
        )

        new_row['lag_30'] = (
            state_df.iloc[-30]['sales']
            if len(state_df) >= 30 else last_row['sales']
        )


        new_row['rolling_mean_4'] = (
            state_df['sales'].tail(4).mean()
        )

        new_row['rolling_std_4'] = (
            state_df['sales'].tail(4).std()
        )


        new_row['month'] = future_date.month
        new_row['quarter'] = future_date.quarter
        new_row['weekofyear'] = future_date.isocalendar().week


        new_row['category_encoded'] = (
            last_row['category_encoded']
        )


        temp_df = pd.DataFrame([new_row])


        X_future = temp_df[[
            'lag_1',
            'lag_7',
            'lag_30',
            'rolling_mean_4',
            'rolling_std_4',
            'month',
            'quarter',
            'weekofyear',
            'category_encoded'
        ]]

        prediction = xgb_model.predict(X_future)[0]

        new_row['sales'] = prediction


        state_df = pd.concat([
            state_df,
            pd.DataFrame([new_row])
        ])

        future_predictions.append({
            'date': future_date,
            'state': state,
            'predicted_sales': prediction
        })



forecast_df = pd.DataFrame(future_predictions)

forecast_df.head()
```

```python
%%writefile app.py

from fastapi import FastAPI
import pandas as pd
import joblib


model = joblib.load('best_model.pkl')


forecast_df = pd.read_csv(
    'future_8weeks_forecast.csv'
)


app = FastAPI(
    title="Sales Forecast API",
    description="Forecast next 8 weeks and single prediction API",
    version="1.0"
)


@app.get('/')

def home():

    return {
        'message': 'Sales Forecast API Running'
    }


@app.get('/forecast/{state}')

def get_forecast(state: str):

    result = forecast_df[
        forecast_df['state'].str.lower()
        == state.lower()
    ]

    if result.empty:

        return {
            'error': 'State not found'
        }

    output = result.to_dict(
        orient='records'
    )

    return {
        'state': state,
        'next_8_weeks_forecast': output
    }



@app.post('/predict')

def predict(

    lag_1: float,
    lag_7: float,
    lag_30: float,
    rolling_mean_4: float,
    rolling_std_4: float,
    month: int,
    quarter: int,
    weekofyear: int,
    category_encoded: int

):

    input_data = pd.DataFrame([{

        'lag_1': lag_1,
        'lag_7': lag_7,
        'lag_30': lag_30,
        'rolling_mean_4': rolling_mean_4,
        'rolling_std_4': rolling_std_4,
        'month': month,
        'quarter': quarter,
        'weekofyear': weekofyear,
        'category_encoded': category_encoded

    }])

    prediction = model.predict(input_data)[0]

    return {

        'predicted_sales': float(prediction)

    }
```
